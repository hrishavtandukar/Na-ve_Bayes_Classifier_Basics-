```python
# importing/loading the Pandas libraries as 'pd' 
import pandas as pd 
data = pd.read_excel("data.xlsx")
```

```python
data
```

```python
data['Possible_to_Play'] = data['Possible to Play'].apply(lambda x:1 if x=="Yes" else 0)
data
```

```python
#creating a dictionary for unique values of a column and giving it an ordinal values
temperature_dict = dict(zip(data['Temperature'].unique(),range(len(data['Temperature'].unique())))) 
temperature_dict 
```

```python
#applying lambda function in order to assign ordinal values to a column
data['Temperature_Ordinal'] = data['Temperature'].apply(lambda x: temperature_dict[x])
```

```python
data
```

```python
#creating a dictionary for unique values of a column and giving it an ordinal values
weather_dict = dict(zip(data['Weather'].unique(),range(len(data['Weather'].unique())))) 
weather_dict 
```

```python
#applying lambda function in order to assign ordinal values to a column
data['Weather_Ordinal'] = data['Weather'].apply(lambda x: weather_dict[x])
data
```

```python
label=list(data['Possible_to_Play'])
label
```

```python
#creating a function which returns weather and temperature in the form of tuple
def return_weather_temperature_tuple(weather_ord,temperature_ord):
    return weather_ord,temperature_ord
```

```python
features=[]
for i in range(len(data['Weather_Ordinal'])):
    features.append(return_weather_temperature_tuple(data['Weather_Ordinal'][i],data['Temperature_Ordinal'][i]))
features
```

```python
# importing Gaussian Naive Bayes model from sklearn.naive_bayes
from sklearn.naive_bayes import GaussianNB
```

```python
# creating a Gaussian Classifier
model = GaussianNB()
```

```python
# training the model using the training sets
model.fit(features,label)
```

```python
predicted= model.predict([[0,2]]) # 0:Sunny, 2:Mild
print ("Predicted Value (Can go out to play):", "yes" if predicted==1 else "no")
```
